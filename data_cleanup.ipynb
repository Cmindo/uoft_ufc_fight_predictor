```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Import UFC Match History Data

ufc = pd.read_csv('ufc_data.csv')
ufc.head(7)
```

```python
ufc.describe()
```

```python
# Drop features

ufc = ufc.drop(columns=['Referee','location','weight_class','B_total_rounds_fought','R_total_rounds_fought'], inplace=False)
```

```python
# Due to rule changes in 2000, all matches prior this year will be dropped as this can affect our model

ufc['date'] = pd.to_datetime(ufc['date'])
ufc['year'] = ufc['date'].dt.year
ufc = ufc[ufc['year'] > 1999]
ufc.head(5)
```

```python
# Update existing features as required

title_bout = []
for index, row in ufc.iterrows():
    if row['title_bout'] == True:
        title_bout.append('1')
    else:
        title_bout.append('0')
ufc['title_bout'] = title_bout
```

```python
# Create deltas for features

ufc['delta_age'] = ufc['R_age']-ufc['B_age']
ufc['delta_Weight_lbs'] = ufc['R_Weight_lbs']-ufc['B_Weight_lbs']
ufc['delta_Reach_cms'] = ufc['R_Reach_cms']-ufc['B_Reach_cms']
ufc['delta_total_time_fought_seconds'] = ufc['R_total_time_fought(seconds)']-ufc['B_total_time_fought(seconds)']
ufc['delta_avg_TOTAL_STR_att'] = ufc['R_avg_TOTAL_STR_att']-ufc['B_avg_TOTAL_STR_att']
ufc['delta_avg_TOTAL_STR_landed'] = ufc['R_avg_TOTAL_STR_landed']-ufc['B_avg_TOTAL_STR_landed']
ufc['delta_avg_SIG_STR_att'] = ufc['R_avg_SIG_STR_att']-ufc['B_avg_SIG_STR_att']
ufc['delta_avg_SIG_STR_landed'] = ufc['R_avg_SIG_STR_landed']-ufc['B_avg_SIG_STR_landed']
ufc['delta_avg_TD_att'] = ufc['R_avg_TD_att']-ufc['B_avg_TD_att']
ufc['delta_avg_TD_landed'] = ufc['R_avg_TD_landed']-ufc['B_avg_TD_landed']
ufc['delta_avg_opp_SIG_STR_att'] = ufc['R_avg_opp_SIG_STR_att']-ufc['B_avg_opp_SIG_STR_att']
ufc['delta_avg_opp_SIG_STR_landed'] = ufc['R_avg_opp_SIG_STR_landed']-ufc['B_avg_opp_SIG_STR_landed']
ufc['delta_avg_opp_TD_att'] = ufc['R_avg_opp_TD_att']-ufc['B_avg_opp_TD_att']
ufc['delta_avg_opp_TD_landed'] = ufc['R_avg_opp_TD_landed']-ufc['B_avg_opp_TD_landed']
ufc['delta_avg_opp_TOTAL_STR_att'] = ufc['R_avg_opp_TOTAL_STR_att']-ufc['B_avg_opp_TOTAL_STR_att']
ufc['delta_avg_opp_TOTAL_STR_landed'] = ufc['R_avg_opp_TOTAL_STR_landed']-ufc['B_avg_opp_TOTAL_STR_landed']
ufc['delta_avg_BODY_att'] = ufc['R_avg_BODY_att']-ufc['B_avg_BODY_att']
ufc['delta_avg_BODY_landed'] = ufc['R_avg_BODY_landed']-ufc['B_avg_BODY_landed']
ufc['delta_avg_GROUND_att'] = ufc['R_avg_GROUND_att']-ufc['B_avg_GROUND_att']
ufc['delta_avg_GROUND_landed'] = ufc['R_avg_GROUND_landed']-ufc['B_avg_GROUND_landed']
ufc['delta_avg_HEAD_att'] = ufc['R_avg_HEAD_att']-ufc['B_avg_HEAD_att']
ufc['delta_avg_HEAD_landed'] = ufc['R_avg_HEAD_landed']-ufc['B_avg_HEAD_landed']
ufc['delta_current_win_streak'] = ufc['R_current_win_streak']-ufc['B_current_win_streak']
ufc['delta_current_lose_streak'] = ufc['R_current_lose_streak']-ufc['B_current_lose_streak']
ufc['delta_win_by_KO/TKO'] = ufc['R_win_by_KO/TKO']-ufc['B_win_by_KO/TKO']
ufc['delta_win_by_Submission'] = ufc['R_win_by_Submission']-ufc['B_win_by_Submission']

ufc.head(5)
```

```python
# Introduce dependent column for Logistic Regression (i.e. Won)

Won = []
for index, row in ufc.iterrows():
    if row['Winner'] == 'Red':
        Won.append('1')
    else:
        Won.append('0')
        
ufc['Won'] = Won
ufc[['R_fighter','B_fighter','Winner','Won']].head(5)
```

```python
#Set Features sets here

new_schema=['Won','title_bout','R_fighter','B_fighter','year','Winner','delta_age','delta_Weight_lbs','delta_Reach_cms','delta_total_time_fought_seconds','delta_avg_SIG_STR_att','delta_avg_SIG_STR_landed','delta_avg_TD_att','delta_avg_TD_landed','delta_avg_opp_SIG_STR_att','delta_avg_opp_SIG_STR_landed','delta_avg_opp_TD_att','delta_avg_opp_TD_landed','delta_avg_opp_TOTAL_STR_att','delta_avg_opp_TOTAL_STR_landed','delta_avg_BODY_att','delta_avg_BODY_landed','delta_avg_GROUND_att','delta_avg_GROUND_landed','delta_avg_HEAD_att','delta_avg_HEAD_landed']
new_schema2=['Won','title_bout','R_fighter','B_fighter','year','Winner','delta_age','delta_Weight_lbs','delta_Reach_cms','delta_total_time_fought_seconds','delta_avg_SIG_STR_att','delta_avg_SIG_STR_landed','delta_avg_TD_att','delta_avg_TD_landed','delta_avg_opp_SIG_STR_att','delta_avg_opp_SIG_STR_landed','delta_avg_opp_TD_att','delta_avg_opp_TD_landed','delta_avg_opp_TOTAL_STR_att','delta_avg_opp_TOTAL_STR_landed']
new_schema3=['Won','title_bout','R_fighter','B_fighter','year','Winner','delta_age','delta_Weight_lbs','delta_Reach_cms','delta_total_time_fought_seconds','delta_avg_SIG_STR_att','delta_avg_SIG_STR_landed','delta_avg_TD_att','delta_avg_TD_landed','delta_avg_opp_SIG_STR_att','delta_avg_opp_SIG_STR_landed','delta_avg_opp_TD_att','delta_avg_opp_TD_landed','delta_avg_opp_TOTAL_STR_att','delta_avg_opp_TOTAL_STR_landed','delta_current_win_streak','delta_current_lose_streak','delta_win_by_KO/TKO','delta_win_by_Submission']
new_schema4=['Won','title_bout','R_fighter','B_fighter','year','Winner','delta_age','delta_Weight_lbs','delta_Reach_cms','delta_total_time_fought_seconds','delta_avg_TOTAL_STR_att','delta_avg_TOTAL_STR_landed','delta_avg_SIG_STR_att','delta_avg_SIG_STR_landed','delta_avg_TD_att','delta_avg_TD_landed','delta_avg_opp_SIG_STR_att','delta_avg_opp_SIG_STR_landed','delta_avg_opp_TD_att','delta_avg_opp_TD_landed','delta_avg_opp_TOTAL_STR_att','delta_avg_opp_TOTAL_STR_landed']
new_schema5=['Won','title_bout','R_fighter','B_fighter','year','Winner','delta_age','delta_Weight_lbs','delta_Reach_cms','delta_total_time_fought_seconds','delta_avg_SIG_STR_att','delta_avg_SIG_STR_landed','delta_avg_TD_att','delta_avg_TD_landed','delta_avg_opp_SIG_STR_att','delta_avg_opp_SIG_STR_landed','delta_avg_opp_TD_att','delta_avg_opp_TD_landed','delta_avg_opp_TOTAL_STR_att','delta_avg_opp_TOTAL_STR_landed']

#Adjust Schema here
ufc_cleansed = ufc[new_schema5].sort_values('year', ascending=False)
ufc_cleansed.reindex(columns=ufc_cleansed['year'])
ufc_cleansed.head(5)
```

```python
# Import fighter-style data and merge with both fighters

R_style = pd.read_csv('R_names_classified_styles.csv')
ufc_cleansed_R_style = ufc_cleansed.merge(right=R_style, how='left', on='R_fighter')

B_style = pd.read_csv('B_names_classified_styles.csv') 
ufc_cleansed_RB_style = ufc_cleansed_R_style.merge(right=B_style, how='left', on='B_fighter')
```

```python
# Drop Null data

ufc_cleansed_RB_style_nonullstats = ufc_cleansed_RB_style[ufc_cleansed_RB_style['delta_total_time_fought_seconds'].notnull()]
ufc_cleansed_RB_style_nonullstats_nullRstyle = ufc_cleansed_RB_style_nonullstats[ufc_cleansed_RB_style_nonullstats['R_Wrestling'].notnull()]
ufc_cleansed_RB_style_nonullstats_nullRBstyle = ufc_cleansed_RB_style_nonullstats_nullRstyle[ufc_cleansed_RB_style_nonullstats_nullRstyle['B_Wrestling'].notnull()]
ufc_cleansed_RB_style_nonullstats_nullRBstyle.head(5)
```

```python
# Export Data, checkpoint

ufc_cleansed_RB_style_nonullstats_nullRBstyle.to_csv(path_or_buf='C:/Users/ChristianMindo/ufc_cleansed_RB_style_nonullstats_nullRBstyle.csv', float_format='%.2f', index=False)
```

```python
# Re-import data then create deltas for fighter style data

final = pd.read_csv('ufc_cleansed_RB_style_nonullstats_nullRBstyle.csv')
final.head(5)
```

```python
# Create deltas for fighter style data

final['delta_Wrestling'] = final['R_Wrestling']-final['B_Wrestling']
final['delta_Jiu-Jitsu'] = final['R_Jiu-Jitsu']-final['B_Jiu-Jitsu']
final['delta_Kickboxing'] = final['R_Kickboxing']-final['B_Kickboxing']
final['delta_Boxing'] = final['R_Boxing']-final['B_Boxing']
final['delta_Karate'] = final['R_Karate']-final['B_Karate']
final['delta_Judo'] = final['R_Judo']-final['B_Judo']
final['delta_MMA'] = final['R_MMA']-final['B_MMA']
```

```python
# Remove original fighter style columns

remove_columns = ['year','Winner','R_fighter','B_fighter','R_Wrestling','R_Jiu-Jitsu','R_Kickboxing','R_Boxing','R_Karate','R_Judo','R_MMA','B_Wrestling','B_Jiu-Jitsu','B_Kickboxing','B_Boxing','B_Karate','B_Judo','B_MMA']
final.drop(columns=remove_columns, inplace=True)
final.dropna(how='any', inplace=True)
```
